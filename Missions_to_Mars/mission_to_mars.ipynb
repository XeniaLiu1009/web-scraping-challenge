## Part1: Scraping

```python
# load the packages
from splinter import Browser
import requests
import os
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

## NASA  Mars News

```python
# set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False) # headless = True you will not see the browser
```

```python
# visit the website using browser
url = "https://redplanetscience.com"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html,'html.parser')
print(soup.prettify())
```

```python
# collec the latest time 
new_time = soup.find_all("div",class_="list_date")[0].text
print(new_time)
```

```python
# collect the latest news title
new_title = soup.find_all("div",class_="content_title")[0].text
print(new_title)

```

```python
# collect the latest news paragraph
new_p = soup.find_all("div",class_="article_teaser_body")[0].text
print(new_p)
```

```python
# kill the browser
browser.quit()
```

## Mars Hemispheres

```python
# set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False) # headless = True you will not see the browser
```

```python
# visit the website using browser
url4 = "https://marshemispheres.com"
browser.visit(url4)
```

```python
html4 = browser.html
soup4 = BeautifulSoup(html4,'html.parser')
print(soup4.prettify())
```

```python
queries = soup4.find_all("div",class_ = "item")
print(queries[0].prettify())
```

```python
# title = queries[0].h3.text
# title
```

```python
# image = queries[0].img["src"]
# image_url = "https://marshemispheres.com/" + image
# image_url
```

```python
titles = []
img_urls = []
for i in range(len(queries)):
    titles.append(queries[i].h3.text)
    img_urls.append("https://marshemispheres.com/" + queries[i].img["src"])

hemisphere_image_urls = []
for title, img_url in zip(titles,img_urls):
    hemisphere_image_url = {'title':title,'img_url':img_url}
    hemisphere_image_urls.append(hemisphere_image_url)
hemisphere_image_urls
```

```python
# print the result
print("item in hemisphere_image_urls : ")
for index, item in enumerate(hemisphere_image_urls):
    print(f"item{index} has values {item['title']} and {item['img_url']}")
```

```python
# kill the browser
browser.quit()
```
